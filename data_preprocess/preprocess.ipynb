```python
import numpy as np
import csv

# 加载.npz文件
data_img = np.load('k49-train-imgs.npz')
data_label = np.load('k49-train-labels.npz')
```

```python

# 提取图像数组
labels = data_label['arr_0']
images = data_img['arr_0']
```

```python
# 将图像数组展平成一维数组
flattened_images = images.reshape(images.shape[0], -1)

# 创建并写入 CSV 文件
with open('char_images_train.csv', mode='w', newline='') as file:
    writer = csv.writer(file)

    # 写入表头，每个像素位置作为一个特征列
    header = ['pixel' + str(i) for i in range(flattened_images.shape[1])]
    header.append('label')
    writer.writerow(header)

    # 写入数据
    for row in images:
        writer.writerow(row)
```

```python

with open('label.csv', mode='w', newline='') as file:
    writer = csv.writer(file)

    # 写入表头，每个像素位置作为一个特征列
    header = ["label"]
    writer.writerow(header)

    # 写入数据
    for label in labels:
        writer.writerow([label])
```
